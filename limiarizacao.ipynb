. Limiarização: a imagem após o limiar é mais nítida visualmente. O limiar de imagem transforma uma imagem em uma imagem binária para extrair informações.

    - É uma forma de segmentar a imagem em tons de cinza em uma imagem binária.

    - Um valor específico de intensidade de pixel é considerado um valor limite.

    - Todos os pixels maiores ou menores que o valor limite são atribuídos ao valor máximo ou mínimo.

    - Ele converte a imagem inteira em uma imagem binária, porque agora existem apenas dois pixels.

    Exemplo: Suponha que queiramos aplicar limiar de imagem aos valores de intensidade de pixel 123, 50, 180, 200. E definimos o valor limite para 128. Portanto, todos os valores maiores que 128 serão o valor de intensidade de pixel mais alto de 255 (brancos), e menores que 128 serão 0 (pretos).

. Limiarização global: Quando o mesmo processo ou valor limite é aplicado à imagem inteira.

. Limiarização local/adaptativa: Funciona localmente, podemos aplicar diferentes valores de limiar para diferentes segmentos da imagem.

Limiar Simples 

    -> Limiar global, precisamos definir um valor de intensidade limite (limiar).

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('teste.webp')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
th_val, binary = cv2.threshold(gray,120,255,cv2.THRESH_BINARY)
#120 é o valor de limite binário e 255 é o valor máximo de intensidade. Portanto, todos os valores de intensidade menores do que 120 serão definidos 0, e todos os valores maiores do que 120 serão definidos 255.
```

```python
plt.imshow(binary, cmap='gray')
```

Limiar inverso binário

    -> O inverso do limiar binário. Definimos o valor limite como 120 e o valor máximo como 255. Portanto, todos os valores de intensidade menores que 120 serão 255 (branco) e 0 (preto) caso contrário.

```python
th_val, binary_inv = cv2.threshold(gray,120,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(binary_inv, cmap='gray')
```

Limiar truncado

    -> No exemplo, definimos o valor limite 145 e o valor máximo de 255. Portanto, para limite truncado, todos os valores menores que 145 serão definidos para o valor limite de 145. Cao contrário, permanecerá o mesmo.

```python
th_val,thr_trunc = cv2.threshold(gray,145,255,cv2.THRESH_TRUNC)
plt.imshow(thr_trunc,cmap='gray')
#São muito mais visíveis do que nas técnicas anteriores.
```

Limiar para zero

    -> Os valores de pixel menores que o limite serão definidos como zero e permanecerão o mesmo caso contrário.

```python
th_val,thr_zero = cv2.threshold(gray,145,255,cv2.THRESH_TOZERO)
plt.imshow(thr_zero,cmap='gray')
```

Limiar para zero inverso

    -> É exatamente o oposto do limite zero.

```python
th_val,thr_zero_inv = cv2.threshold(gray,145,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thr_zero_inv,cmap='gray')
```
